#### General Klein-Nishina cross section

```python
def sigma_kn(x):
    """
    Total Klein-Nishina cross section,
    normalised by Thomson cross section
    """
    s = 3./4 * (1 + x) / x**3 * (2 * x * (1 + x) / (1 + 2 * x) - numpy.log(1 + 2 * x))
    s += numpy.log(1 + 2 * x) / (2 * x) - (1 + 3 * x)/(1 + 2 * x)**2
    return s
```
